# Esercitazione 5

In questa esercitazione torno ad usare la classe che ho preparato per il calcolo delle medie a blocchi. La funzione *CalcolaBlocco* questa volta deve ottenere il valore medio di distribuzioni di probabilità usando l'algoritmo di Metropolis. Per utilizzare l'algoritmo di Metropolis prima devo scegliere una probabilità di transizione $T(\vec{x}|\vec{y})$, in modo da riuscire a proporre una nuova configurazione $\vec{x}$, e poi calcolare l'accettazione data da

$$
A(\vec{x}|\vec{y}) = min\bigg(1, \frac{T(\vec{y}|\vec{x})p(\vec{x})}{T(\vec{x}|\vec{y})p(\vec{y})}\bigg)
$$

dove $p(x)$ è la densità di probabilità che voglio campionare. Comincio considerando una probabilità di transizione uniforme, in questo modo la forma dell'accettazione si semplifica e diventa $A(\vec{x}|\vec{y}) = min(1, p(\vec{x})/p(\vec{y}))$. Il metodo *CalcolaBlocco* deve quindi usare $T$ per ottenere una nuova configurazione e poi accettare o meno questa con probabilità data da $A$. 

## Funzioni d'onda e passo Metropolis

L'esercizio dà come distribuzioni di probabilità da campionare due moduli quadri di funzioni d'onda, ovvero 

$$
|\Psi_{1,0,0}(r,\theta,\phi)|^2=
\frac{a_0^{-3}}{\pi} e^{-2r/a_0} 
$$

$$
|\Psi_{2,1,0}(r,\theta,\phi)|^2=
\frac{a_0^{-5}}{32 \pi}r^2 e^{-r/a_0} \cos^2(\theta)
$$

dove $a_0$ è il raggio di Bohr. Il metodo *CalcolaBlocco* parte da una data posizione nello spazio e trova una nuova configurazione aggiungendo una variabile casuale distribuita uniformemente tra $[-a, a]$ ad ogni coordinata, dove $a$ è la distanza massima del passo Metropolis e deve essere scelta in modo da avere una accettazione di circa il $50\%$. Dopo aver trovato una nuova posizione, questa viene accettata secondo $A$. Ripetendo questo procedimento molte volte posso ottenere il valore medio della distribuzione campionata.

Nella simulazione utilizzo come unità della distanza il raggio di Bohr $a_0$. Dopo aver fatto più simulazioni ho trovato che per avere un'accettazione del $50\%$ per $|\Psi_{1,0,0}|^2$ devo usare un passo di $1.2$, mentre per $|\Psi_{2,1,0}|^2$ un passo di lunghezza $3$.

## Dimensione dei blocchi

Una cosa da stabilire prima di guardare i risultati di una simulazione è la dimensione dei blocchi. Questi devono essere sufficientemente grandi da eliminare le correlazioni dovute al passo finito dell'algoritmo. Guardando la forma delle distribuzioni di probabilità, in questo caso posso stimare la dimensione dei blocchi senza dover calcolare una funzione di autocorrelazione: le due distribuzioni decadono esponenzialmente quindi è molto difficile ottenere valori della distanza.

Ad esempio diciamo che dalla prima distribuzione è molto difficile ottenere delle distanze superiori ai cinque raggi di Bohr. Posso quindi pensare in prima approssimazione di avere una distribuzione non nulla soltanto in una sfera di raggio $5 a_0$. Per stimare quanto può essere grande l'effetto dell'autocorrelazione, mi chiedo quanti passi Metropolis siano necessari per andare da un qualsiasi punto dentro questa sfera a un altro qualsiasi punto. Considerando che in questo caso ogni step Metropolis può spostare al massimo di $1.2$ in ogni direzione carteziana, segue che nei casi migliori sono sufficienti qualche passo Metropolis, nei casi peggiori non servono più di $100$ step. Di conseguenza un blocco che comprende $1000$ passi Metropolis dovrebbe essere già sufficiente. In modo analogo si ragiona per la seconda distribuzione di probabilità.

## Equilibrazione

Il discorso precedente funziona nel caso in cui il mio sistema abbia raggiunto l'equilibrio. Se parto da un punto ragionevole, ovvero dove la densità di probabilità non è troppo piccola, non ho bisogno di equilibrare. Questo avviene perchè la posizione di partenza in questo caso è una configurazione che potrei avere ottenuto una volta raggiunto l'equilibrio. Se invece la mia simulazione comincia da una posizione molto lontana dall'origine, come possono essere $100 a_0$, allora chiaramente i primi blocchi avranno dei valori medi del raggio molto alti e quindi poco realistici. Inoltre a grande distanza un passo Metropolis non cambia significativamente il raggio e quindi la probabilità del punto attuale e del nuovo punto proposto è molto simile portanto a un'accettazione molto alta.

## Simulazioni

Per le seguente simulazioni ho utilizzato come punto di partenza $(x, y, z) = (1, 1, 1)$ che corrisponde a $r = \sqrt{3}$ in unità del raggio di Bohr. In questo modo non ho bisogno di equilibrare perchè questo punto potrebbe facilmente essere un punto ottenuto una volta raggiunta l'equilibrazione. Utilizzo $100$ blocchi, ognuno con $10^5$ passi Metropolis.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carica i dati
x_unif100, unif100, err_unif100 = np.loadtxt("../Esercizio5/uniforme_100.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
x_unif210, unif210, err_unif210 = np.loadtxt("../Esercizio5/uniforme_210.txt", usecols=(0,1,2), delimiter=' ', unpack='true')

# Prepara due immagini
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# Grafico per psi_100
ax1.errorbar(x_unif100, unif100, yerr = err_unif100, label = "Simulazione")
ax1.axhline(y = 3/2, xmin = 0.05, xmax = 0.95, color = 'red', ls = "--", label = "Valore esatto")
ax1.set_xlabel('Numero metropolis')
ax1.set_ylabel(r'$\langle r \rangle \enspace (a_0)$')
ax1.set_title(r"Distanza media di $\Psi_{1,0,0}}$ con transizione uniforme")
ax1.legend()
ax1.grid(True)

# Grafico per psi_210
ax2.errorbar(x_unif210, unif210, yerr = err_unif210, label = "Simulazione")
ax2.axhline(y = 5, xmin = 0.05, xmax = 0.95, color = 'red', ls = "--", label = "Valore esatto")
ax2.set_xlabel('Numero metropolis')
ax2.set_ylabel(r'$\langle r \rangle \enspace (a_0)$')
ax2.set_title(r"Distanza media di $\Psi_{2,1,0}}$ con transizione uniforme")
ax2.legend()
ax2.grid(True)
```

Si vede che le simulazioni tendono al valore atteso. Posso anche controllare che il Metropolis stia campionando correttamente il modulo quadro delle due funzioni d'onda. Per fare questo registro la posizione ogni $10$ passi Metropolis e faccio una simulazione con blocchi di $1000$ step per non avere troppi punti.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Carica i dati
X100, Y100, Z100 = np.loadtxt("../Esercizio5/pos_unif_100.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
X210, Y210, Z210 = np.loadtxt("../Esercizio5/pos_unif_210.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
```

```python
# Disegno i punti ottetuti con Psi_100
fig = plt.figure(figsize=(15,5), tight_layout = True)
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(X100, Y100, Z100, c=Z100, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(r"$\Psi_{1,0,0}}$ campionata con Metropolis")
ax.view_init(10, 30)

# Disegno i punti ottenuti con Psi_210
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(X210, Y210, Z210, c=Z210, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(r"$\Psi_{2,1,0}}$ campionata con Metropolis")
ax.view_init(10, 30)
```

Oltre a un grafico in 3D che non è mai facilmente interpretabile, nel caso della prima funzione d'onda posso anche controllare facilmente la distribuzione radiale, questa è proporzionale a $r^2|\Psi_{1,0,0}|^2$, dove la $r^2$ è dovuto alle coordinate sferiche. Nel caso della seconda distribuzione non posso fare lo stesso procedimento perchè questa dipende anche dall'angolo.

```python
# Calcola la distanza radiale
R = X100**2 + Y100**2 + Z100**2
R = np.sqrt(R)

# Fai un istogramma con le distanze
n_bins = 50
fig, ax = plt.subplots(1, 1,figsize=(7,5))
ax.set_xlabel('Raggio')
ax.set_ylabel('Probabilità')
ax.set_title(r'Distribuzione radiale di $\Psi_{1,0,0}}$')
n, bins, path = ax.hist(R, n_bins, density = True, range = (0,6), label = "Simulazione")
ax.grid(True)

# Plotta la distribuzione esatta
x = np.linspace(0, 6, 100)
ax.plot(x, x**2 * np.exp(-2*x) * 4, label = "Distribuzione esatta")
ax.legend()
plt.show()
```

La distribuzione ottenuta con il Metropolis concorda con quella esatta. 

## Transizione gaussiana

Le stesse simulazioni si possono fare modificando la probabilità di transizione, adesso al posto di una probabilità uniforme utilizzo una gaussiana. Il procedimento per calcolare il valore medio del raggio è analogo, l'unica modifica è che devo modificare il passo per mantenere l'accettazione al $50\%$. In generale devo diminuire la lunghezza del passo perchè questo adesso non rappresenta più la lunghezza massima di spostamento, ma la deviazione standard della gaussiana. Le seguenti simulazioni sono state ottenute utilizzando $100$ blocchi, ognuno con $10^5$ step Metropolis e passo $0.8$ per $|\Psi_{1,0,0}|^2$ e $1.8$ per $|\Psi_{2,1,0}|^2$.

```python
# Carica i dati
x_gauss100, gauss100, err_gauss100 = np.loadtxt("../Esercizio5/gaussiana_100.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
x_gauss210, gauss210, err_gauss210 = np.loadtxt("../Esercizio5/gaussiana_210.txt", usecols=(0,1,2), delimiter=' ', unpack='true')

# Prepara due immagini
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# Grafico per psi_100
ax1.errorbar(x_gauss100, gauss100, yerr = err_gauss100, label = "Simulazione")
ax1.axhline(y = 3/2, xmin = 0.05, xmax = 0.95, color = 'red', ls = "--", label = "Valore esatto")
ax1.set_xlabel('Numero metropolis')
ax1.set_ylabel(r'$\langle r \rangle \enspace (a_0)$')
ax1.set_title(r"Distanza media di $\Psi_{1,0,0}}$ con transizione gaussiana")
ax1.legend()
ax1.grid(True)

# Grafico per psi_210
ax2.errorbar(x_gauss210, gauss210, yerr = err_gauss210, label = "Simulazione")
ax2.axhline(y = 5, xmin = 0.05, xmax = 0.95, color = 'red', ls = "--", label = "Valore esatto")
ax2.set_xlabel('Numero metropolis')
ax2.set_ylabel(r'$\langle r \rangle \enspace (a_0)$')
ax2.set_title(r"Distanza media di $\Psi_{2,1,0}}$ con transizione gaussiana")
ax2.legend()
ax2.grid(True)
```

Anche in questo caso le simulazioni corcordano con il valore esatto essendo entro una deviazione standard di distanza. Inoltre confontando le incertezze ottenute con probabilità di transizione diverse, si nota che queste sono comparabili.

## Partenza lontano dall'origine

Voglio provare a far partire la simulazione da un punto lontano dall'origine per far vedere come varia il risultato. Considero soltanto la prima distribuzione e una transizione uniforme senza cambiare la lunghezza del passo $1.2$. La seguente simulazione è stata ottenuta usando $100$ blocchi di lunghezza $10^4$ ciascuno e usando come punto iniziale la posizione $(100, 100, 100)$ in raggi di Bohr.

```python
# Carica i dati
x_unif100, unif100, err_unif100 = np.loadtxt("../Esercizio5/uniforme_100.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
X100, Y100, Z100 = np.loadtxt("../Esercizio5/pos_unif_100.txt", usecols=(0,1,2), delimiter=' ', unpack='true')

# Grafico per psi_100
fig = plt.figure(figsize=(15,5), tight_layout = True)
ax = fig.add_subplot(1, 2, 1)
ax.errorbar(x_unif100, unif100, yerr = err_unif100, label = "Simulazione")
ax.axhline(y = 3/2, xmin = 0.05, xmax = 0.95, color = 'red', ls = "--", label = "Valore esatto")
ax.set_xlabel('Numero metropolis')
ax.set_ylabel(r'$\langle r \rangle \enspace (a_0)$')
ax.set_title(r"Distanza media di $\Psi_{1,0,0}}$ con transizione gaussiana")
ax.legend()
ax.set_ylim(1.4,3)
ax.grid(True)

# Disegno i punti ottetuti con Psi_100
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(X100, Y100, Z100, c=Z100, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(r"$\Psi_{1,0,0}}$ campionata con Metropolis")
ax.view_init(10, 30)
```

Dallo scatter plot si vede che la posizione è partita da lontano e poi si è sempre più avvicinata all'origine finchè non ha raggiuto una zona stabile dove poi è rimasta. Aver raggiunto questa zona corrisponde ad aver completato l'equilibrazione. Nel grafico di sinistra invece si vede l'andamento del valore medio del raggio; questo tende verso il valore corretto ma molto lentamente perchè soprattutto il primi blocchi hanno contribuito con un valore decisamente maggiore rispetto al valore corretto.
