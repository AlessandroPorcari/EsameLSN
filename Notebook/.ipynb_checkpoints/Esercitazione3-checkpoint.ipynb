# Esercizitazione 3

In questa lezione ho usato la classe per le medie a blocchi come nelle lezioni precedenti. La funzione *CalcolaBlocco* deve calcolare il prezzo dell'offerta Call o Put. Per ottenere il costo iniziale della Call devo prima calcolare il guadagno ottenuto alla fine del processo. Questo lo ottengo come $max(0, S(T)-K)$, dove $K$ è il prezzo accordato e $S(T)$ è la cifra che ho alla fine del processo. Il prezzo della Call è ottenuto prendendo il guadagno e scontandolo della quantità che avrei ottenuto decidendo di usare una banca al posto di una opzione Call, ovvero 

$$C[S(0), 0] =\langle max\big(0, S(T)-K\big) e^{-rT}\rangle$$ 

dove $r$ è il tasso di interesse. Il metodo *CalcolaBlocco* segue questo procedimento applicandolo più volte in modo da riuscire a calcolare la media come valore di un blocco. Nel caso dell'offerta Put il procedimento è analogo solo che ora il guadagno è dato da $max(0, K - S(T))$ e di conseguenza il prezzo iniziale diventa

$$P[S(0), 0] = \langle max\big(0, K-S(T)\big) e^{-rT}\rangle$$

Adesso rimane da determinare come ottenere $S(T)$. Il calcolo può essere affrontato in tre modi differenti: il primo consiste nell'utilizzare la soluzione analitica di Black e Scholes, il secondo in una simulazione per ottenere direttamente $S(T)$ e il terzo consiste in una divisione del secondo metodo in step intermedi $S(t_i)$ che portano al valore finale. Il primo metodo è implementato in un seguente specchietto python e viene usato per controllare la correttezza degli altri due metodi. Usando il secondo approccio si ha

$$ S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right] $$

dove $\sigma$ è la volatilità e $W(T)$ è una variabile distribuita in modo gaussiano con media nulla e varianza $T$. Per il calcolo del costo dividendo in step si usa la formula 

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$

dove $Z_i$ è una variabile distribuita in modo gaussiano con media nulla e varianza unitaria.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import scipy
from scipy.special import erf
from numpy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

Ora comincio a fare le simulazioni per il costo iniziale dell'opzione Call/Put. Nei grafici seguenti sono riportati sia il calcolo diretto sia quello diviso in step per entrambe le offerte. Ho usato $100$ blocchi ognuno con $10^4$ calcoli dell'opzione per ogni simulazione.

```python
# Carica i dati
x_callDir, callDir, callDir_err = np.loadtxt("../Esercizio3/call_diretto.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
x_callDis, callDis, callDis_err = np.loadtxt("../Esercizio3/call_discreto.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
x_putDir, putDir, putDir_err = np.loadtxt("../Esercizio3/put_diretto.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
x_putDis, putDis, putDis_err = np.loadtxt("../Esercizio3/put_discreto.txt", usecols=(0,1,2), delimiter=' ', unpack='true')

# Prepara vettori per contenere posizioni, valori ed errori
pos = [x_callDir, x_callDis, x_putDir, x_putDis]
avg = [callDir, callDis, putDir, putDis]
err = [callDir_err, callDis_err, putDir_err, putDis_err]

# Prepara quattro immagini
fig, ax = plt.subplots(2, 2, figsize = (14, 8), tight_layout = True)
ax = ax.reshape(-1)

for i in range(4):
# Grafico per la Call calcolata direttamente
    ax[i].errorbar(pos[i], avg[i], yerr = err[i], label = "Simulazione")
    ax[i].set_xlabel('Simulazioni prezzo offerta')
    ax[i].set_ylabel('Costo')
    ax[i].grid(True)
    if i < 2:
        ax[i].axhline(y = call, xmin = 0.05, xmax = 0.95, color = 'red', ls = "--", label = "Valore esatto")
    else:
        ax[i].axhline(y = put, xmin = 0.05, xmax = 0.95, color = 'red', ls = "--", label = "Valore esatto")
    ax[i].legend()
    
# Scrivi i titoli
ax[0].set_title("Costo finale Call calcolato direttamente")
ax[1].set_title("Costo finale Call calcolato in modo discreto")
ax[2].set_title("Costo finale Put calcolato direttamente")
ax[3].set_title("Costo finale Put calcolato in modo discreto")
plt.show()
```

Tutti i risultati ottenuti dalle simulazioni sono compatibili con la soluzione analitica di Black e Scholes, infatti tutti i risultati distano meno di una deviazione standard dal valore esatto. Inoltre si ottiene che l'incertezza finale ottenuta con il metodo diretto è della stessa grandezza di quella ottenuta con il metodo discreto.
