# Esercitazione 9

In questa esercitazione devo risolvere il problema del commesso viaggiatore utilizzando un algoritmo genetico. Come codice ho deciso di creare una classe *Cromosoma*. Questa classe contiene un vettore per l'ordine delle città, una variabile per la lunghezza del percorso e delle funzioni che permettono di calcolare la lunghezza del percorso e applicare delle mutazioni.

## Generazione condizione iniziale

Il programma usa come parametri il numero delle città, il numero delle generazioni da fare, la popolazione per generazione e un bool che indica se le città vengono generate su una circonferenza o meno. Dopo questo genera le posizioni delle città e genera in modo casuale le sequenze iniziali. Per generare la posione di una città prendo un angolo a caso in $[0,2\pi]$ oppure due punti casuali $x, y \in [0,1]$ a seconda che io voglia le città sulla circonferenza o su un quadrato. Per generare in modo casuale una sequenza comincio creando un vettore ordinato, successivamente scambio ogni elemento (tranne il primo) con un altro elemento a caso del vettore, ovvero se ho $N$ città allora per ogni $i \in [2,N]$ prendo a caso un $j \in [2,N]$ e scambio $i$ con $j$.

Dopo avere generato le sequenze iniziali devo calcolare la lunghezza del percorso associato, per fare questo chiamo il metodo *Cost* che calcola la lunghezza secondo $L^{(1)}$. La generazione delle prime sequenze e il calcolo della loro lunghezza sono raccolte in un unico metodo *GeneraCromosomi*. Il prossimo passo è ordinare la popolazione in base alla lunghezza in modo da semplificare la selezione. Come algoritmo di ordinamento utilizzo un semplice Selection Sort. 

## Selezione

Generate le condizioni iniziali posso cominciare con l'algoritmo genetico. Faccio un primo ciclo sulle generazioni e un secondo ciclo per generare tutte gli individui della nuova generare. Ho deciso che una nuova generazione contiene solo nuovi individui, cioè non ho applicato nessuna forma di elitismo. Siccome i migliori individui hanno già una probabilità maggiore di essere scelti per riprodursi, di conseguenza è relativamente facile che costruendo una nuova popolazione capiti che vengano prese le sequenze migliori e non vengano applicati crossover e mutazioni. Questo implica che è probabile che i migliori individui siano presenti nella nuova generazione anche senza elitismo. 

Per generare due sequenze della nuova generazione devo prima selezionarne due. Per la selezione ho deciso di prendere la seguenza in posizione $k$ secondo $p(k) \propto \frac{1}{k}$. Creo un vettore che contiene le probabilità di selezione di ogni posizione in modo da non dovere calcolare le probabilità ogni volta che chiamo la selezione. La funzione *Selezione* sceglie a caso due elementi distinti dalla generazione corrente secondo la probabilità detta sopra.

## Crossover

Dopo aver scelto due sequenze l'algoritmo prova ad applicare il crossover con una certa probabilità. Se il Crossover non viene applicato allora vado direttamente alle Mutazioni con le due sequenze date dalla selezione. Altrimenti scelgo a caso un indice $i \in [3, N-1]$ e comincio a preparare le due sequenze figlie: un figlio contiene i primi $i-1$ elementi di una delle due sequenze data dalla selezione. Per completare il primo figlio guardo l'altra sequenza data dalla selezione: prendo il primo elemento della sequenza, se questo non è nel primo figlio lo aggiungo, altrimenti passo al secondo elemento e così via. Il secondo figlio si ottiene in maniera analoga soltanto scambiando i ruoli delle due sequenze genitori. In questo modo si capiscono i limiti posti sull'indice i: non può essere $1$ perchè il primo elemento è sempre lo stesso e quindi non cambierebbe niente, non può essere $2$ oppure $N$ perchè altrimenti i figli coincidono con i genitori.

## Mutazioni

Per avere le nuove sequenze che faranno parte della nuova generazione devo provare ad applicare le mutazioni. Ho deciso di introdurre quattro mutazioni diverse. Chiamando il metodo *Mutazioni*, provo ad applicare ognuna delle quattro diverse mutazioni con la stessa probabilità. Tutte le mutazioni non modificano il primo indice, di conseguenza nelle sequenti descrizioni non lo specificherò.

La prima mutazione consiste nel prendere due elementi diversi e scambiarli. La seconda mutazione scambia due blocchi di elementi della stessa dimensione. Questo implica che la dimensione del blocco deve essere minore di $N/2$, altrimenti non riesco a prendere due blocchi distinti sufficientemente grandi. In questa mutazione come anche nelle successive ho deciso di utilizzare delle condizioni periodiche. In questo modo non devo limitare la dimensione del blocco in base a dove prendo il primo elemento. Le condizioni periodiche permettono di avere una maggiore mobilità degli elementi agli estremi che altrimenti verrebbero modificati meno volte rispetto ai centrali. Le condizioni periodiche non considerano il primo elemento, quindi dopo quello in posizione $N$ si passa direttamente a quello in posizione $2$.

La terza mutazione prende un certo numero di elementi contigui e inverte la loro posizione. Il numero degli elementi da invertire deve essere compreso tra $2$ e $N-1$. Un caso particolare si verifica quando si vogliono invertire $N-1$ elementi. Se come elemento di partenza di prende il primo della sequenza, l'ultimo della sequenza oppure l'elemento centrale (se dispari, altrimenti i due centrali) allora la sequenza che si ottiene è uguale a quella di partenza, ma soltanto invertita. Questi casi non modificano la lunghezza del percorso, ma ho deciso di lasciarli perchè sono abbanstanza rari e possono permette di esplorare soluzioni diverse con crossover.

L'ultima mutazione prende un blocco di elementi e li sposta avanti di una certa quantità. Gli elementi incontrari spostando il blocco vengono retrocessi. Se sposto tutti i geni questo comporta soltanto una traslazione di tutti gli elementi di un certo valore. La dimensione del blocco può essere presa tra $1$ e $N-1$. La scelta del numero di spostamenti è invece più delicata: con $N-1$ spostamenti tornerei alla configurazione iniziale. Inoltre quando il numero degli spostamenti sommato alla dimensione del blocco raggiunge $N$, il risultato di questa mutazione è una traslazione di tutti gli elementi. Questa particolarità succede spesso, infatti essendo gli spostamenti compresi in $[1, N-2]$ e la dimensione del blocco in $[1,N-1]$, segue che le possibili configurazioni sono $(N-1)(N-2)$, mentre quelle che sommate fanno almeno $N$ sono $\frac{(N-1)(N-2)}{2}$, esattamente la metà. La quarta mutazione di conseguenza metà delle volte che viene chiamata trasla solamente tutti gli indici. Per limitare questo caso diminuisco il valore massimo degli spostamenti in modo da avere un traslazione completa solo nel caso in cui il blocco da spostare contiene tutti gli elementi.

Dopo aver applicato le mutazioni calcolo la lunghezza dei due percorsi ottenuti e li aggiungo alla nuova generazione. Quando la popolazione della nuova generazione ha raggiunto il numero stabilito, sostituisco la vecchia generazione con la nuova e la ordino in base alla lunghezza. Alla fine di ogni generazione stampo su file la lunghezza del percorso migliore e la media della metà migliore della popolazione per verificare l'andamento dell'algoritmo genetico.

## Simulazione

Nella simulazione seguente ho disposto le città su una circonferenza, ho usato $32$ città, $200$ generazioni ognuna con una popolazione di $1000$ individui. Come probabilità del crossover ho usato $0.7$ e come probabilità di applicare una qualsiasi mutazione $0.1$.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carica i dati
x, y = np.loadtxt("../Esercizio9/posizioni_città.txt", usecols=(0,1), delimiter=' ', unpack='true')
gen, minimo, media = np.loadtxt("../Esercizio9/generazioni.txt", usecols=(0,1,2), delimiter=' ', unpack='true')

# Prepara due immagini
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6), tight_layout=True)

# Disegna il percorso migliore
ax1.plot(x, y, '-o')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title("Percorso più breve")
ax1.grid(True)

# Plotta la lunghezza più corta ogni generazione
ax2.plot(gen, minimo, label = "Lunghezza minore")
ax2.plot(gen, media, label = "Media su metà popolazione")
ax2.set_xlabel('Generazioni')
ax2.set_ylabel('Lunghezza')
ax2.set_title("Lunghezza del percorso al variare della generazione")
ax2.set_xlim(0,50)
ax2.legend(loc = (0.5,0.8))
ax2.grid(True)
```

La soluzione trovata è facilmente la migliore, questo lo posso dire perchè la soluzione ottimale deve essere simile a una circonferenza. Dall'andamento del lunghezza si vede che l'algoritmo generico ha impiegato circa $30$ generazioni per trovare il risultato ottimale. La media sulla metà migliore della popolazione dice che si raggiunge molto velocemente una situazione in cui quasi tutti gli individui contengono la stessa sequenza. Questo implica che i possibili miglioramenti della soluzione trovata si possono ottenere soltanto tramite le mutazioni. Per evitare che tutti gli elementi della popolazione diventino uguali posso aumentare la probabilità di mutazione. Potrei anche provare a cambiare leggermente la probabilità di selezione, ma ho notato che questo sposta soltanto di qualche generazione la convergenza alla soluzione unica.
