## Simulated annealing

In questo esercizio devo risolvere lo stesso problema della scorsa esercitazione ma utilizzando un simulated annealing. In questo caso non ho una popolazione di sequenze, ma ne ho soltanto una con cui esploro le possibili configurazioni tramite Metropolis. Questo significa che non utilizzo più le funzioni per la selezione e per il crossover. La mutazioni adesso diventano il modo per proporre una mossa del Metropolis, di conseguenza le devo applicare a ogni passo della simulazione.

Ho deciso di modifare leggermente il metodo *Mutazioni*: prima provava ad applicare ogni mutazione con una certa probabilità, adesso a ogni passo del Metropolis applico soltanto una a caso delle quattro mutazioni. Questa decisione è stata fatta perchè devo essere sicuro di applicarne almeno una ogni step, ma dopo aver fatto delle simulazioni ho notato che applicando più di una mutazione ogni step il risultato peggiorava. L'algoritmo di Metropolis applica una mutazione alla sequenza, se la sequenza mutata ha una lunghezza più corta di quella da cui partito allora la prendo come nuova sequenza, altrimenti la accetto con probabilità $ p(x) = exp\big((y-x)/T\big)$ dove $T$ è la temperatura del sistema, $y$ è la lunghezza della mia sequenza, $x$ è la lunghezza della sequenza proposta.


Il simulated annealing funziona abbassando la temperatura in modo da far raggiungere al sistema una condizione di minimo di energia. Prima di fare una simulazione devo quindi decidere come fare variare la temperatura. Ho deciso di dividere la temperatura per una costante $K$ ogni volta che devo cambiare temperatura, questo è stato scelto perchè ho visto che forniva risultati corretti. Data una temperatura iniziale $T_i$, una finale $T_f$ e il numero delle temperature che voglio simulare $N$, la costante è data da $K = \sqrt[N-1]{\frac{T_i}{T_f}}$.

## Simulazione

La seguente simulazione è stata fatta utilizzando $T_i = 5$, $T_f = 0.01$, utilizzando $20$ temperature, $10^5$ passi Metropolis per temperatura e le città distribuite su una circonferenza.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carica i dati
x, y = np.loadtxt("../Esercizio10/Esercizio10.1/posizioni_città.txt", usecols=(0,1), delimiter=' ', unpack='true')
x_minimo, minimo = np.loadtxt("../Esercizio10/Esercizio10.1/costo.txt", usecols=(0,1), delimiter=' ', unpack='true')

# Prepara due immagini
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6), tight_layout=True)

# Disegna il percorso migliore
ax1.plot(x, y, '-o')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title("Percorso più breve")
ax1.grid(True)

# Plotta la lunghezza più corta ogni generazione
ax2.plot(x_minimo, minimo)
ax2.set_xlabel('Passi Metropolis')
ax2.set_ylabel('Lunghezza')
ax2.set_title("Percorso più breve al variare dei passi Metropolis")
ax2.grid(True)
```

Si vede che il risultato è uguale a quello ottenuto con l'algoritmo genetico. Inoltre dal percorso più breve si trova che la soluzione ottimale è stata ottenuta intorno a $1.34$ $10^6$ passi Metropolis, questo coincide con la quattordicesima temperatura, ovvero circa $T = 0.51$. Adesso possiamo fare la stessa simulazione con gli stessi parametri per le città in un quadrato. 

```python
# Carica i dati
x, y = np.loadtxt("../Esercizio10/Esercizio10.1/posizioni_città.txt", usecols=(0,1), delimiter=' ', unpack='true')
x_minimo, minimo = np.loadtxt("../Esercizio10/Esercizio10.1/costo.txt", usecols=(0,1), delimiter=' ', unpack='true')

# Prepara due immagini
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6), tight_layout=True)

# Disegna il percorso migliore
ax1.plot(x, y, '-o')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title("Percorso più breve")
ax1.grid(True)

# Plotta la lunghezza più corta ogni generazione
ax2.plot(x_minimo, minimo)
ax2.set_xlabel('Passi Metropolis')
ax2.set_ylabel('Lunghezza')
ax2.set_title("Percorso più breve al variare dei passi Metropolis")
ax2.grid(True)
```

## Calcolo parallelo

In questa seconda parte dell'esercitazione devo utilizzare lo stesso algoritmo genetico utilizzato per la lezione precedente, l'unica differenza è che adesso devo far partire quattro processi contemporaneamente che si scambiano informazione. Come prima cosa devo aggiungere due variabili che mi specificano ogni quante generazioni i processi si devo scambare informazione e quante sequenze si scambiano ogni volta.

La prossima cosa da modificare è il generatore di numeri casuali, questo perchè altrimenti tutti i processi prendeno lo stesso seme per il generatore e quindi danno risultati uguali. Risolvere questo problema è facile, infatti basta modificare la funzione *generatore* per fare in modo che il seme che prende dipenda dal rank del processo. Il prossimo problema da risolvere è la generazione delle città, questo perchè avendo diversi generatori di numeri casuali avrò diverse posizioni delle città e quindi i processi risolverebbero problemi indipendenti. Per ovviare a questo faccio generare le città soltanto a un processo e poi faccio un broadcast per mandare le posizioni agli altri processi.

Per avere uno scambio casuale tra i vari processi, faccio generare a un singolo processo un numero casuale intero nell'intervallo $[1,3]$. Questo mi indica con quale altro processo deve scambiare informazione. Successivamente trasmetto questa scelta con un broadcast. Il processo che è stato scelto scambierà informazione con quello che ha inviato il broadcast, mentre i due rimanenti si invieranno informazioni tra di loro. Per lo scambio uso le funzioni *Isend* e *Recv*, in questo modo non avendo un invio bloccante possono subito cominciare a ricevere evitando di riempire il buffer. 

## Simulazione

Per la seguente simulazione ho utilizzato gli stessi parametri che utilizzavo nell'esercizio precedente ovvero $32$ città, $200$ generazioni con una popolazione di $1000$ sequenze, una probabilità di crossover di $0.7$ e una probabilità per ogni mutazione di $0.1$. Decido di scambiare soltanto l'individuo migliore ogni $30$ generazioni. La simulazione la faccio soltanto per il caso delle città nel quadrato. Nel seguente grafico si ha il risultato del primo processo.

```python
# Carica i dati
x, y = np.loadtxt("../Esercizio10/Esercizio10.2/posizioni_città0.txt", usecols=(0,1), delimiter=' ', unpack='true')
gen, minimo, media = np.loadtxt("../Esercizio10/Esercizio10.2/generazioni0.txt", usecols=(0,1,2), delimiter=' ', unpack='true')

# Prepara due immagini
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6), tight_layout=True)

# Disegna il percorso migliore
ax1.plot(x, y, '-o')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title("Percorso più breve")
ax1.grid(True)

# Plotta la lunghezza più corta ogni generazione
ax2.plot(gen, minimo, label = "Lunghezza minore")
ax2.plot(gen, media, label = "Media su metà popolazione")
ax2.set_xlabel('Generazioni')
ax2.set_ylabel('Lunghezza')
ax2.set_title("Lunghezza del percorso al variare della generazione")
ax2.set_xlim(20,130)
ax2.set_ylim(4.5,6.5)
ax2.legend(loc = (0.5,0.8))
ax2.grid(True)
```

Dalla figura del percorso si vede che questa è identica a quella ottenuta con il simulated annealing. Guardando l'andamento della lunghezza in funzione delle generazioni si vede che anche in questo caso si raggiunge velocemente una situazione stabile. Tuttavia si osserva che in corrispondenza degli scambi (ovvero alle generazioni $30, 60, 90, 120$) si hanno delle diminuizioni della lunghezza. Questo significa che scambiare l'individuo migliore con un altro processo ha aiutato a raggiungere una soluzione migliore.
