In questa lezione devo calcolare la funzione d'onda dello stato fondamentale con metodi variazionali. Per fare questo devo calcolare il sequente integrale

$$
\langle {\hat H} \rangle_T = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

Per il calcolo di questo integrale utilizzo la prima frazione come densità di probabilità da cui campionare i punti e la seconda frazione come integranda. Per riuscire a ottenere dei punti generati secondo quella distribuzione di probabilità utilizzo l'algoritmo di Metropolis con probabilità di transizione uniforme. Al denominatore della densità da campionare ho soltanto una normalizzazione che poi viene semplificata nel Metropolis, di conseguenza l'accettazione del nuovo punto sarà dettata soltanto dalla forma del modulo quadro della funzione d'onda, ovvero

$$|\Psi_T(x)|^2 \propto e^{-\frac{x^2}{\sigma^2}}\bigg(1 + cosh\frac{2 \mu x}{\sigma^2}\bigg)$$

## Calcolo dell'integranda

Per poter fare la simulazione adesso devo calcolare il termine 

$$
\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)} = \frac{{\big(-\frac{\hbar^2}{2 m} \frac{\partial^2}{\partial x^2}} + V(x)\big) \Psi_T(x)}{\Psi_T(x)} \quad, \quad \Psi_T(x) \propto e^{-\frac{(x - \mu)^2}{2 \sigma^2}} + e^{-\frac{(x + \mu)^2}{2 \sigma^2}}$$

Per il calcolo della derivata seconda si ha

$$\frac{\partial^2}{\partial x^2} \Psi_T(x) \propto \frac{1}{\sigma^2} \bigg[e^{-\frac{(x-\mu)^2}{2\sigma^2}}\bigg(\frac{(x-\mu)^2}{\sigma^2}-1\bigg)+e^{-\frac{(x+\mu)^2}{2\sigma^2}}\bigg(\frac{(x+\mu)^2}{\sigma^2}-1\bigg)\bigg] = \\
\frac{1}{\sigma^2} \bigg[-\Psi_T(x) + e^{-\frac{(x-\mu)^2}{2\sigma^2}}\frac{(x-\mu)^2}{\sigma^2}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}\frac{(x+\mu)^2}{\sigma^2}\bigg] = \\
\frac{1}{\sigma^4} \bigg[\big((x+\mu)^2-\sigma^2\big)\Psi_T(x) - 4 \mu x e^{-\frac{(x-\mu)^2}{2\sigma^2}}\bigg]$$

E quindi la funzione da calcolare diventa

$$ \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)} = \frac{\hbar^2}{2m\sigma^4} \bigg[\sigma^2 - (x+\mu)^2 + \frac{4 \mu x}{1 + e^{-\frac{2 \mu x}{\sigma^2}}}\bigg] + x^4 - \frac{5}{2}x^2$$

Per l'equilibrazione se parto da un punto con probabilità non bassa, non c'è bisogno che io la faccia perchè potrebbe essere un punto ottenuto una volta già raggiunta l'equilibrazione. ## Calcolo dell'energia con parametri ottimali

La seguente simulazione è stata ottenuta utilizzando i parametri ottimali, un passo di $2.5$ e $100$ blocchi da $10^5$ step Metropolis.

```python
# Carica i dati 
x_ene, ene, ene_err =  np.loadtxt("../Esercizio8/energia.txt", usecols=(0,1,2), delimiter=' ', unpack='true')

# Plotta i risultati
fig, ax = plt.subplots(1, 1, figsize = (7, 5))
ax.errorbar(x_ene, ene, yerr = ene_err)
ax.set_xlabel("Passi Metropolis")
ax.set_ylabel("Energia")
ax.set_title("Energia dello stato $\Psi^{\mu=0.8}_{\sigma=0.62}$")
ax.grid(True)
```

Il valore ottenuto può essere confrontato con il valore dato dalla trasformazione dell'equazione di Shroedinger in una equazione matriciale. Questo calcolo viene svolto nel seguente specchietto python.

```python
# Potenziale
def Vpot(x):
    return (x**2 - 2.5) * x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a / 2, a / 2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2
CDiff = np.diag(np.ones(N - 1), -1) - 2 * np.diag(np.ones(N), 0) + np.diag(np.ones(N - 1), 1)

# Hamiltonian matrix
H = (-(hbar**2) * CDiff) / (2 * m * dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E, psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi / np.sqrt(dx)

print("Ground state energy: ", E[0])
```

Come mi aspettavo il valore trovato con il metodo variazionale è correttamente maggiore di quello trovato con il problema matriciale. La differenza tra i due risultati è minore di $0.02$, quindi abbiamo trovato una buona approssimazione per la funzione d'onda.

## Andamento della funzione d'onda

Posso paragonare la forma della funzione d'onda con i nostri parametri ottimizzati e quella ottenuta con il metodo matriciale. Allo stesso tempo posso anche controllare che la distribuzione campionata corrisponda effettivamente al modulo quadro della funzione d'onda di prova. Per il seguente grafico ho salvato la posizione ogni $10$ passi e ho usato $10$ blocchi ognuno con $10^5$ passi Metropolis.

```python
# Per normalizzazione funzione d'onda di prova
import scipy.integrate as integrate
```

```python
# Modulo quadro funzione d'onda di prova
def Psi2(x, mu, sigma):
    return np.exp(-x**2 / sigma**2) * (1 + np.cosh(2 * mu * x / sigma**2))

# Carica i dati dell'istogramma
pos = np.loadtxt("../Esercizio8/posizioni.txt", unpack='true')

# Disegna la curva ottenuta con le matrici
fig, ax = plt.subplots(1, 1, figsize = (7, 5))
ax.plot(x, psi[0]**2, c = "black", label = "Matrici")
ax.set_title("Modulo quadro funzioni d'onda fondamentali")
ax.set_xlabel("x")
ax.set_ylabel("Probabilità")
ax.set_xlim((-3,3))
ax.set_ylim((0,0.5))

# Disegna la funzione di prova e l'istogramma con le posizioni
psi_trial = Psi2(x, mu = 0.8, sigma = 0.62)
norma = integrate.quad(Psi2, -5, 5, args = (0.8, 0.62))
ax.plot(x, psi_trial / norma[0], c = "r", label = "Variazionale")
ax.hist(pos, bins = 50, density = True, label = "Campionato")
ax.legend()
ax.grid(True)
```

Dal grafico si vede che i punti campionati seguono la distribuzione di probabilità che abbiamo utilizzato per il metodo variazionale. La funzione variazionale rispetto a quella matriciale ha i due massimi più alti mentre ha il minimo più basso. Inoltre si nota anche che la funzione variazionale ha i due massimi a valori leggermente minori. Nella parte in cui la funzione d'onda decade le due curve sono molto simili.
