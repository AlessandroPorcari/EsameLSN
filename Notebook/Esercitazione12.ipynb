## Rete neurale profonda

In questa esercitazione devo costruire delle reti capaci di riconoscere una cifra presente in un'immagine. Nella prima richiesta viene già fornito il codice che svolge questa operazione, io devo soltanto modificare l'ottimizzatore per provare a ottenere dei risultati migliori. In questo caso utilizzo una rete neurale che contiene quattro layer: il primo ha $400$ neuroni e utilizza una attivazione ReLU, il secondo $100$ neuroni e una ReLU, il terzo è un Dropout e il quarto contiene $10$ neuroni con una attivazione softmax in modo da avere soltanto $10$ possibili esiti. La simulazione utilizza $10$ epoche per allenare la rete neurale. Ho deciso di provare a utilizzare come ottimizzatori AdaGrad, Nadam e Adamax. Nei seguenti specchietti python è implementata questa rete neurale.

```python
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.datasets import mnist

seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# Dimensione delle immagini caricate e numero dei possibili risultati finali
img_rows, img_cols = 28, 28
num_classes = 10

# Carica i dati dividendo in training e verifica
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# Riscala i dati nell'intervallo [0,1]
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

# Converti in vettori per usare categorical cross entropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import optimizers
from tensorflow.keras.optimizers import SGD, Adagrad, Adamax, Nadam

def create_DNN():
    
    # Costruisci il modello
    model = Sequential()
    model.add(Dense(400,input_shape=(img_rows*img_cols, ), activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    
    # Compila il modello
    model.compile(loss = keras.losses.categorical_crossentropy,
                  optimizer = Nadam(), metrics = ['acc'])
    
    return model
```

```python
# Crea la rete neurale e fai il fit
model_DNN = create_DNN()
history = model_DNN.fit(X_train, Y_train,
          batch_size = 32, epochs = 10,
          verbose=1, validation_data=(X_test, Y_test))
```

```python
# Prepara quattro grafici
fig, ax = plt.subplots(2, 2, figsize=(14,8), tight_layout = True)
ax = ax.reshape(-1)
histories = [history_sgd, history_adagrad, history_nadam, history_adamax]
nomi = ["SGD", "AdaGrad", "Nadam", "Adamax"]

# Grafici per accuracy
for i in range(2):
    for j in range(2):
        ax[i].plot(histories[j + 2*i].history['acc'], label = "Train " + nomi[j + 2*i])
        ax[i].plot(histories[j + 2*i].history['val_acc'], label = "Test " + nomi[j + 2*i])
    ax[i].set_ylabel('Accuracy')
    ax[i].set_xlabel('Epoche')
    ax[i].set_title("Accuracy del DNN con " + nomi[2*i] + " e " + nomi[2*i + 1])
    ax[i].legend(loc = "best")
    ax[i].grid()

# Grafici per loss
for i in range(2):
    for j in range(2):
        ax[i + 2].plot(histories[j + 2*i].history['loss'], label = "Train " + nomi[j + 2*i])
        ax[i + 2].plot(histories[j + 2*i].history['val_loss'], label = "Test " + nomi[j + 2*i])
    ax[i + 2].set_ylabel('Loss')
    ax[i + 2].set_xlabel('Epoche')
    ax[i + 2].set_title("Loss del DNN con " + nomi[2*i] + " e " + nomi[2*i + 1])
    ax[i + 2].legend(loc = "best")
    ax[i + 2].grid()
```

Nella prima immagine vediamo che l'accuratezza dei dati di validazione è maggiore rispetto a quella dei dati di training, questo è dovuto alla presenza del Dropout che viene applicato soltanto nella fase di allenamento. Se non usassi il layer di Dropout allora otterrei un'accuratezza migliore sui dati di training. Una considerazione analoga vale per l'andamento della funzione loss ottenuta utilizzando lo stochastic gradient descent e AdaGrad. Dall'andamento della funzione di loss e della accuratezza possiamo affermare che SGD porta in questo caso a una rete neurale migliore rispetto a AdaGrad.

Nei grafici con Nadam e Adamax notiamo che la loss dei dati di training riesce a diventare più bassa di quella dei dati di validazione anche se è presente un layer di Dropout. Inoltre vediamo che questi due ottimizzatori portano a un risultato migliore rispetto allo stochastic gradient descent. Da questa simulazione non riesco però a determinare quale sia il miglior ottimizzatore, infatti Nadam presenta una loss minore sui dati di training ma maggiore sui dati di validazione.

## Rete neurale convoluzionale

Per utilizzare una rete neurale convoluzionale devo cambiare la forma del mio input. In questo caso non comincio più da una configurazione lineare come nel caso della rete precedente, ma da una immagine in due dimensioni e di conseguenza devo fare un reshape dei miei dati iniziali. Tale operazione lo faccio con il seguente specchietto python.

```python
# Fai un reshape per utilizzare una rete convoluzionale
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
```

Prima di fare una simulazione devo scegliere la struttura dei layer della rete. Ho deciso di utilizzare nove layer: i primi due sono un layer convoluzionale con otto filtri e un MaxPooling che concentra l'informazione contenuta in ogni filtro. I prossimi due layer sono nuovamente uno convoluzionale e un MaxPooling, soltanto che adesso utilizzo sedici filtri. 

A questo punto voglio cominciare a utilizzare dei layer densi, per fare ciò introduco un Flatten. Successivamente aggiungo due layer densi rispettivamente da $200$ e $100$ neuroni, poi un layer di Dropout e infine un ultimo layer denso con $10$ neuroni che mi restituisce il valore della cifra letta. Tutti i layer che necessitano di una funzione di attivazione utilizzano una ReLU tranne l'ultimo che usa una softmax. Come ottimizzatore ho deciso di utilizzare lo stochastic gradient descent. In questa simulazion utilizzo $5$ epoche.

```python
from tensorflow.keras.layers import Flatten, Conv2D, MaxPooling2D

def create_CNN():
    
    # Costruisci il modello
    model = Sequential()
    model.add(Conv2D(8, kernel_size=(5, 5), activation='relu', input_shape = input_shape))
    model.add(MaxPooling2D(pool_size = (2, 2)))
    model.add(Conv2D(16, kernel_size=(5, 5), activation='relu'))
    model.add(MaxPooling2D(pool_size = (2, 2)))
    model.add(Flatten())
    model.add(Dense(200))
    model.add(Dense(100))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    
    # Compila il modello
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='SGD', metrics=['acc'])
    
    return model
```

```python
# Crea la rete neurale e fai il fit
model_CNN = create_CNN()
history = model_CNN.fit(X_train, Y_train,
          batch_size = 32, epochs = 5,
          verbose=1, validation_data = (X_test, Y_test))
```

```python
# Prepara due grafici
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5), tight_layout = True)

# Grafico per accuracy
ax1.plot(history.history['acc'])
ax1.plot(history.history['val_acc'])
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Epoche')
ax1.set_title("Accuracy della rete neurale convoluzionale")
ax1.legend(['Train', 'Test'], loc='best')
ax1.grid()

# Grafico per loss
ax2.plot(history.history['loss'])
ax2.plot(history.history['val_loss'])
ax2.set_ylabel('Loss')
ax2.set_xlabel('Epoche')
ax2.set_title("Loss della rete neurale convoluzionale")
ax2.legend(['Train', 'Test'], loc='best')
ax2.grid()
```

Dai grafici vediamo che questa rete neurale ha raggiungo una accuratezza del $97\%$ per i dati di training e del $98\%$ per quelli di validazione, valori paragonabili con quelli ottenuti dai migliori ottimizzatori della DNN. Il piccolo valore della funzione loss e l'elevata correttezza delle previsioni ci fa affermanre che questa rete neurale riesce a prevedere in modo corretto il valore presente sull'immagine.

## Immagini scritte a mano

Abbiamo verificato che le neurali prodotte funzionano bene sui dati di training e di validazione. Ora voglio controllare che continuino a dare risultati corretti anche quando lavorano su delle immagini che non hanno mai visto. Per fare questa verifica preparo $10$ immagini di $28\times28$ pixel con sfondo nero e cifra bianca e le passo alle reti neurali prodotte. Come DNN utilizzo quella ottenuta con Nadam e non con lo stochastic gradient descent poichè quest'ultima forniva risultati leggermente peggiori.

```python
from PIL import Image

# Funzione per caricare una immagine
def immagine(num):

    digit_filename = f"./Immagini/{num}.png"
    digit_in = Image.open(digit_filename).convert('L')

    ydim, xdim = digit_in.size
    pix = digit_in.load();
    data = np.zeros((xdim, ydim))
    for j in range(ydim):
        for i in range(xdim):
            data[i,j]=pix[j,i]

    return data
```

```python
# Carica le immagini
immagini = np.array([immagine(i) for i in range(10)])

# Ottieni la previsione con DNN
immagini = immagini.reshape(immagini.shape[0], img_rows * img_cols)
predictionsDNN = model_DNN.predict(immagini)

# Ottieni la previsione con CNN
immagini = immagini.reshape(immagini.shape[0], img_rows, img_cols, 1)
predictionsCNN = model_CNN.predict(immagini)

# Plotta i risultati
plt.figure(figsize = (15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(immagini[i, :, :, 0], cmap = 'gray')    
    plt.title("DNN: {} \nCNN: {}".format(np.argmax(predictionsDNN[i]), np.argmax(predictionsCNN[i])))    
    plt.axis('off') 
```

La rete neurale convoluzionale è riuscità a identificare correttamente tutti i numeri, mentre la DNN ha sbagliato in corrispondenza del $7$. Questo conferma il fatto che le reti neurali prodotte funzionano in modo corretto e riescono a identificare le cifre in modo efficiente. Volendo ottenere delle reti più precise potrei provare ad aumentare il numero delle epoche poichè in questo caso ne ho usate poche.
